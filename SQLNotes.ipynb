<h3 style = "color:green">9. Window Functions

<div style = "background-color:lightblue; color:green; font-size:20px; text-align:justify">Window functions let us perform calculations across a set of rows related to the current row, without collapsing rows like GROUP BY does.

```python
%%sql

SELECT employees.sex, MAX(employees.salary) AS 'highest_salary'
FROM employees
JOIN employees_info
    ON employees.id = employees_info.id
GROUP BY sex;
```

<div style = "background-color:lightblue; color:green; font-size:20px; text-align:justify">The output of grouped rows is aggregated into single row.

```python
%%sql

SELECT  sex, MAX(salary) OVER(PARTITION BY sex ORDER BY id) AS 'highest_salary'
FROM employees;
```

```python
%%sql

SELECT first_name, last_name, sex, salary,
SUM(salary) OVER(PARTITION BY sex ORDER BY id) AS rolling_total
FROM employees_info;
```

<div style = "background-color:lightblue; color:green; font-size:20px; text-align:justify">
<ul>
<li>SUM() is a window function.</li>
<li>salary is expression, a target column on which window function operates.</li>
<li>OVER clause defines the window partitions to form groups of rows.</li>
<li>PARTITION clause divides the rows into partitions over which window functions applies. If the PARTITION clause is not defined, whole result set is treated as single partition.</li>
<li>ORDER BY clause specifies the order of partitions on which window function applies.</li>
</ul>
</div>

<h3 style = "color:green">9.1 Ranking Window Functions

```python
%%sql

SELECT name, salary, 
    ROW_NUMBER() OVER(ORDER BY salary DESC) AS 'continuous_rank',
    RANK() OVER(ORDER BY salary DESC) AS 'tie_rank',
    DENSE_RANK() OVER(ORDER BY salary DESC) AS 'continuous_tie_rank',
    PERCENT_RANK() OVER(ORDER BY salary DESC) AS 'percent_rank'
FROM employees;
```

```python
%%sql

SELECT name, salary,
    NTILE(4) OVER(ORDER BY salary) AS quartile
FROM employees;
```

<div style = "background-color:lightblue; color:green; font-size:20px; text-align:justify">Divides rows into n equal buckets.

<h3 style = "color:green">9.2 Aggregate Window Functions

```python
%%sql

SELECT name, salary, sex,
    SUM(salary) OVER(PARTITION BY sex ORDER BY salary DESC) AS total_salary_by_sex,
    AVG(salary) OVER(PARTITION BY sex ORDER BY salary DESC) AS average_salary_by_sex,
    MAX(salary) OVER(PARTITION BY sex ORDER BY salary DESC) AS highest_salary_by_sex,
    MIN(salary) OVER(PARTITION BY sex ORDER BY salary DESC) AS lowest_salary_by_sex,
    COUNT(*) OVER(PARTITION BY sex ORDER BY salary DESC) AS total_employees_by_sex
FROM employees;
```

<h3 style = "color:green">9.3 Value/ Offset Functions

<div style = "background-color:lightblue; color:green; font-size:20px; text-align:justify">These look forward or backward at other rows.

```python
%%sql

SELECT name, salary,
    LAG(salary, 1, 0) OVER(ORDER BY salary DESC ) AS values_from_previous_rows,
    salary - LAG(salary, 1, 0) OVER(ORDER BY salary DESC) AS difference_in_salary
FROM employees;
```

```python
%%sql

SELECT name, salary,
    LEAD(salary, 1) OVER(ORDER BY salary DESC) AS salary_from_next_rows
FROM employees;
```

```python
%%sql

SELECT name, salary,
    FIRST_VALUE(name) OVER(ORDER BY salary DESC) AS highest_earner
FROM employees;
```

```python
%%sql

SELECT name, salary,
    NTH_VALUE(name, 2) OVER(ORDER BY salary DESC) AS second_highest_earner
FROM employees;
```

<h3 style = "color:green">9.5 FRAME clause

<div style = "background-color:lightblue; color:green; font-size:20px; text-align:justify">It defines which rows should be included in the window function calculation.

<div style = "background-color:lightblue; color:green; font-size:20px; text-align:justify">
Basic structure:<br>
<pre>
window_function(expression) OVER(
    PARTITION BY...
    ORDER BY...
    ROWS | RANGE BETWEEN frame_start and frame_end


)
</pre>
</div>

<div style = "background-color:lightblue; color:green; font-size:20px; text-align:justify">
<pre>
UNBOUNDED PRECEDING
N-PRECEDING
CURRENT ROW
M-FOLLOWING
UNBOUNDED FOLLOWING
</pre>
</div>

```python
%%sql

SELECT name, salary,
    SUM(salary) OVER(ORDER BY id ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING) AS total_salary
FROM employees;
```

<div style = "background-color:lightblue; color:green; font-size:20px; text-align:justify">If we start from first row(current row), there is no preceding row so it sums the salary from all the following rows. If we go for second row(current row), it sums the preceding row(first row) and all following rows and so on..

```python
%%sql

SELECT name, salary,
    SUM(salary) OVER(ORDER BY id ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) AS running_total
FROM employees;
```

```python
%%sql

SELECT name, salary,
    SUM(salary) OVER(ROWS BETWEEN CURRENT ROW AND UNBOUNDED FOLLOWING) AS running_total
FROM employees; 
```

```python
%%sql

SELECT name, salary,
    SUM(salary) OVER(ROWS BETWEEN 2 PRECEDING AND CURRENT ROW) AS running_total
FROM employees;
```

```python
%%sql

SELECT name, salary,
    SUM(salary) OVER(RANGE BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW ) AS running_total
FROM employees;
```

<div style = "background-color:lightblue; color:green; font-size:20px; text-align:justify">RANGE treats same ORDER BY value as a group.

<h3 style = "color:green">10. Common Table Expressions (CTEs)

<div style = "background-color:lightblue; color:green; font-size:20px; text-align:justify">CTEs allows to define a named, temporary result set that can be referenced within a query. It is like named subquery that lives only for the duration of that query. It makes complex queries much easier to write, read, and maintain.
</div>

<div style = "background-color:lightblue; color:green; font-size:20px; text-align:justify">
Basic Syntax:
<pre>
With cte_name AS(
    subquery
    SELECT...
)
SELECT * FROM cte_name;
</pre>

</div>

```python
# %%sql

# SELECT
#   id,
#   name,
#   salary,
#   SUM(salary) over(order by id) AS rolling_sum
#   FROM employees
# where rolling_sum < 1000000;

#Gives error
```

```python
%%sql

WITH cte AS(
    SELECT id, name, salary,
    SUM(salary) OVER(ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) AS rolling_sum
    FROM employees
)

SELECT * 
FROM cte
WHERE rolling_sum <1000000;
```

```python
%%sql

WITH cte1 AS(
    SELECT id, name, sex
    FROM employees
    WHERE name LIKE "%_____%"
),
cte2 AS(
    SELECT id, salary
    FROM employees
    WHERE salary >50000
)

SELECT * 
FROM cte1
JOIN cte2
    on cte1.id = cte2.id;
```

<h4 style = "color:green">Standardizing Data

```python
%%sql

SELECT first_name, last_name, city, country
FROM customers_dirty
LIMIT 10;
```

```python
%%sql

UPDATE customers_dirty
SET 
    first_name = CONCAT(UPPER(SUBSTRING(first_name, 1, 1)), LOWER(RIGHT(first_name, LENGTH(first_name) - 1)))
    ;
```

```python
%%sql

UPDATE customers_dirty
SET
    last_name = CONCAT(UPPER(SUBSTRING(last_name, 1,1)), LOWER(RIGHT(last_name, LENGTH(last_name) - 1)))
    ;
```

```python
%%sql

UPDATE customers_dirty
SET country = 'USA';
```
